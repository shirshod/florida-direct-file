Creates separate dataframes for minors and adults

### Importing libraries

```python
import pandas as pd
```

### Importing dataset

```python
date_columns = ["BIRTH_DATE", "INIT_ARREST_DT", "CLERK_DT", "OFFENSE_DT", "CAPIAS_DT", "REARREST_DT"]
```

```python
obts_basic = pd.read_csv("data/obts/obts_basic_revised.csv", parse_dates = date_columns)
obts_basic["BIRTH_DATE"] = obts_basic["BIRTH_DATE"].apply(lambda x: x.replace(day = 15))

obts_basic.info()
```

```python
obts_basic
```

### Computing age

```python
obts_basic["AGE"] = round((obts_basic["OFFENSE_DT"] - obts_basic["BIRTH_DATE"]) / pd.Timedelta(days = 365.25),2)

obts_basic
```

### Separating out minors and adults

#### Minors

```python
obts_basic_jr = obts_basic[obts_basic["AGE"] < 18].reset_index(drop = True)

obts_basic_jr
```

#### Adults

```python
obts_basic_sr = obts_basic[obts_basic["AGE"] >= 18].reset_index(drop = True)

obts_basic_sr
```

### Exporting dataframes

```python
obts_basic_jr.to_csv("data/obts/obts_basic_jr.csv", index = False)
obts_basic_sr.to_csv("data/obts/obts_basic_sr.csv", index = False)
```
