Computing average sentences for minors compared to adults

### Importing libraries

```python
import pandas as pd
```

### Importing datasets

```python
data_types = {
    "ID": str,
    "ST_IMPOSED_YR": str,
}
```

```python
minors = pd.read_csv("data/obts/minors_analysissentences.csv", dtype = data_types)
adults = pd.read_csv("data/obts/adults_analysissentences.csv", dtype = data_types)

### filters for misdemeanors and felonies
minors = minors[minors["CT_CHARGE_LEVEL"].isin(["M", "F"])].reset_index(drop = True)
adults = adults[adults["CT_CHARGE_LEVEL"].isin(["M", "F"])].reset_index(drop = True)

### removes juvenile sanctions
minors = minors[minors["CT_ACTION_TAKEN"] != "K"].reset_index(drop = True)
adults = adults[adults["CT_ACTION_TAKEN"] != "K"].reset_index(drop = True)

### removes sentences with no confinement
minors = minors[minors["ST_DAYS"] != 0].reset_index(drop = True)
adults = adults[adults["ST_DAYS"] != 0].reset_index(drop = True)
```

### Analyzing

#### Minors 

```python
minors.info()
```

```python
minors.head()
```

```python
minors["COURT_DESIGNATOR"].value_counts()
```

##### Computing averages

```python
### felonies and misdemeanors overall
avg_days = round(minors["ST_DAYS"].mean())
avg_years = round(avg_days/365, 1)
print("Average sentence (felonies + misdemeanors):", avg_days, "days and", avg_years, "years")

### felonies
felony = minors[minors["CT_CHARGE_LEVEL"] == "F"].reset_index(drop = True)
avg_days = round(felony["ST_DAYS"].mean())
avg_years = round(avg_days/365, 1)
print("Average sentence (felonies):", avg_days, "days and", avg_years, "years")

felony_capital = felony[felony["CT_CHARGE_DEGREE"] == "C"].reset_index(drop = True)
avg_days = round(felony_capital["ST_DAYS"].mean())
avg_years = round(avg_days/365, 1)
print("Average sentence (capital felonies):", avg_days, "days and", avg_years, "years")

felony_life = felony[felony["CT_CHARGE_DEGREE"] == "L"].reset_index(drop = True)
avg_days = round(felony_life["ST_DAYS"].mean())
avg_years = round(avg_days/365, 1)
print("Average sentence (life felonies):", avg_days, "days and", avg_years, "years")

felony_lifecap = felony[felony["CT_CHARGE_DEGREE"] == "P"].reset_index(drop = True)
avg_days = round(felony_lifecap["ST_DAYS"].mean())
avg_years = round(avg_days/365, 1)
print("Average sentence (first-degree, punishable by life felonies):", avg_days, "days and", avg_years, "years")

felony_fst = felony[felony["CT_CHARGE_DEGREE"].isin(["F", "S", "T"])].reset_index(drop = True)
avg_days = round(felony_fst["ST_DAYS"].mean())
avg_years = round(avg_days/365, 1)
print("Average sentence (felonies, not life, capital or punishable by life):", avg_days, "days and", avg_years, "years")

felony_first = felony[felony["CT_CHARGE_DEGREE"] == "F"].reset_index(drop = True)
avg_days = round(felony_first["ST_DAYS"].mean())
avg_years = round(avg_days/365, 1)
print("Average sentence (first-degree felonies):", avg_days, "days and", avg_years, "years")

felony_second = felony[felony["CT_CHARGE_DEGREE"] == "S"].reset_index(drop = True)
avg_days = round(felony_second["ST_DAYS"].mean())
avg_years = round(avg_days/365, 1)
print("Average sentence (second-degree felonies):", avg_days, "days and", avg_years, "years")

felony_third = felony[felony["CT_CHARGE_DEGREE"] == "T"].reset_index(drop = True)
avg_days = round(felony_third["ST_DAYS"].mean())
avg_years = round(avg_days/365, 1)
print("Average sentence (third-degree felonies):", avg_days, "days and", avg_years, "years")

### misdemeanors 
misd = minors[minors["CT_CHARGE_LEVEL"] == "M"].reset_index(drop = True)
avg_days = round(misd["ST_DAYS"].mean())
avg_years = round(avg_days/365, 1)
print("Average sentence (misdemeanors):", avg_days, "days and", avg_years, "years")

misd_first = misd[misd["CT_CHARGE_DEGREE"] == "F"].reset_index(drop = True)
avg_days = round(misd_first["ST_DAYS"].mean())
avg_years = round(avg_days/365, 1)
print("Average sentence (first-degree misdemeanors):", avg_days, "days and", avg_years, "years")

misd_second = misd[misd["CT_CHARGE_DEGREE"] == "S"].reset_index(drop = True)
avg_days = round(misd_second["ST_DAYS"].mean())
avg_years = round(avg_days/365, 1)
print("Average sentence (second-degree misdemeanors):", avg_days, "days and", avg_years, "years")
```
