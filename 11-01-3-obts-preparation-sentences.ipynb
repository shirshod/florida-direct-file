Preparing sentencing dataframe for analysis

### Importing libraries

```python
import pandas as pd
```

#### Importing dataset

```python
data_types = {
    "id": str,
    "st_Imposed_Dt": str,
    "st_Confine": str
}
```

```python
sentence = pd.read_csv("data/obts/obts_sentence.csv", dtype = data_types)
sentence.columns = sentence.columns.str.upper()

sentence.head()
```

```python
sentence.info()
```

### Formating dataframe

```python
def convert(value):
    try:
        return str(int(value))
    except ValueError:
        return "N/A"
```

```python
min_date = pd.to_datetime("NaT") 

def date_convert(value):
    if value == "00000000":
        return min_date
    elif value == "88888888":
        return min_date
    elif value == "0.0":
        return min_date
    elif value == "88888888.0":
        return min_date
    else:
        try:
            return pd.to_datetime(value, format = "%Y%m%d")
        except ValueError:
            return min_date
```

```python
sentence["ST_IMPOSED_DT"] = sentence["ST_IMPOSED_DT"].fillna("00000000").replace("NA", "00000000")
sentence["ST_IMPOSED_DT"] = sentence["ST_IMPOSED_DT"].apply(date_convert)
sentence["ST_IMPOSED_YR"] = sentence["ST_IMPOSED_DT"].dt.year


sentence["ST_CONFINE"] = sentence["ST_CONFINE"].apply(convert)
sentence["ST_CONFINE"] = sentence["ST_CONFINE"].apply(lambda x: x.zfill(14))

sentence["ST_CONFINE"] = sentence["ST_CONFINE"].replace("88888888888888", "N/A")
sentence["ST_CONFINE"] = sentence["ST_CONFINE"].replace("00000000000N/A", "N/A")
sentence["ST_CONFINE"] = sentence["ST_CONFINE"].replace("99999999999998", "LIFE")
sentence["ST_CONFINE"] = sentence["ST_CONFINE"].replace("99999999999999", "DEATH")
sentence["ST_CONFINE"] = sentence["ST_CONFINE"].replace("66666666666666", "JUV")
sentence["ST_CONFINE"] = sentence["ST_CONFINE"].replace("55555555555555", "JUV")

sentence
```

### Separating out data for analysis

#### Life, death and juvenile sentences

```python
st_type = ["N/A", "LIFE", "DEATH", "JUV"]

### filters out life, death and juvenile sentences
sentence_other_1 = sentence[sentence["ST_CONFINE"].isin(st_type)].reset_index(drop = True)
sentence_other_1["ST_PART1"] = sentence_other_1["ST_CONFINE"]
sentence_other_1["ST_PART2"] = sentence_other_1["ST_CONFINE"]
```

#### Fixed duration sentences

```python
sentence_main = sentence[~sentence["ST_CONFINE"].isin(st_type)].reset_index(drop = True)

### creates new columns with min and max sentence periods
sentence_main["ST_PART1"] = sentence_main["ST_CONFINE"].apply(lambda x: x[:7])
sentence_main["ST_PART2"] = sentence_main["ST_CONFINE"].apply(lambda x: x[7:])

sentence_main["ST_PART1"] = sentence_main["ST_PART1"].replace("8888888", "0000000")
sentence_main["ST_PART1"] = sentence_main["ST_PART1"].replace("9999998", "LIFE")
sentence_main["ST_PART1"] = sentence_main["ST_PART1"].replace("9999999", "DEATH")
sentence_main["ST_PART1"] = sentence_main["ST_PART1"].replace("6666666", "JUV")
sentence_main["ST_PART1"] = sentence_main["ST_PART1"].replace("5555555", "JUV")

sentence_main["ST_PART2"] = sentence_main["ST_PART2"].replace("8888888", "0000000")
sentence_main["ST_PART2"] = sentence_main["ST_PART2"].replace("9999998", "LIFE")
sentence_main["ST_PART2"] = sentence_main["ST_PART2"].replace("9999999", "DEATH")
sentence_main["ST_PART2"] = sentence_main["ST_PART2"].replace("6666666", "JUV")
sentence_main["ST_PART2"] = sentence_main["ST_PART2"].replace("5555555", "JUV")
```

```python
sentence_other_2 = sentence_main[sentence_main["ST_PART1"].isin(st_type)].reset_index(drop = True)
sentence_other = pd.concat([sentence_other_1, sentence_other_2], axis=0, ignore_index=True)

sentence_other 
```

```python
sentence_main = sentence_main[~sentence_main["ST_PART1"].isin(st_type)].reset_index(drop = True)
sentence_main = sentence_main[~sentence_main["ST_PART2"].isin(st_type)].reset_index(drop = True)

sentence_main
```

```python
### converts sentence to days
days_1 = sentence_main["ST_PART1"].apply(lambda x: int(x[-2:]))
months_1 = sentence_main["ST_PART1"].apply(lambda x: int(x[3:5]))
years_1 = sentence_main["ST_PART1"].apply(lambda x: int(x[:3]))
days_2 = sentence_main["ST_PART2"].apply(lambda x: int(x[-2:]))
months_2 = sentence_main["ST_PART2"].apply(lambda x: int(x[3:5]))
years_2 = sentence_main["ST_PART2"].apply(lambda x: int(x[:3]))

sentence_main["ST_PART1"] = (years_1*365) + (months_1*30) + days_1
sentence_main["ST_PART2"] = (years_2*365) + (months_2*30) + days_2
```

```python
def check_non_zero(row):
    if row["ST_PART1"] != 0 and row["ST_PART2"] != 0:
        return row["ST_PART2"]
    else:
        return row["ST_PART1"] + row["ST_PART2"]
```

```python
sentence_main["ST_DAYS"] = sentence_main.apply(check_non_zero, axis=1)

### changes position of columns
cols = list(sentence_main.columns)
cols = cols[:3] + [cols[-3], cols[-2], cols[-1]] + cols[3:-3]
sentence_main = sentence_main[cols]

sentence_main
```

### Exporting dataframes

```python
sentence.to_csv("data/obts/sentence_master.csv", index = False)
sentence_main.to_csv("data/obts/sentence_main.csv", index = False)
sentence_other.to_csv("data/obts/sentence_other.csv", index = False)
```
