Analyzing acquittals, juvenile sanctions and youthful offender sentences

### Importing libraries

```python
import pandas as pd
```

### Importing dataset

```python
minors_master = pd.read_csv("data/obts/minors_sentences_master.csv")

columns_to_drop = ["MESSAGE_TYPE", "ARREST_NAME", "FINAL_NAME", "PR_FLST_CHAP", "PR_FLST_SECT", "PR_FLST_SUBSECT",
                  "CT_FLST_CHAP", "CT_FLST_SECT", "CT_FLST_SUBSECT", "ST_PROB_LEN", "ST_COMM_LEN", "ST_FINE",
                   "ST_COURT_COST", "ST_RESTITUTION"]
minors_master = minors_master.drop(columns = columns_to_drop)

minors_master.head()
```

```python
minors_master.info()
```

### Analyzing for acquittals and juvenile sanctions

#### Felonies and misdemeanors

```python
### filters for felonies and misdemeanors
data = minors_master[minors_master["CT_CHARGE_LEVEL"].isin(["M", "F"])].reset_index(drop = True)

### filters for total acquittals
acquit = data[data["CT_ACTION_TAKEN"].isin(["A", "I", "D", "E", "T", "M", "Y"])].reset_index(drop = True)

### filters for acquittals (excluding mental health)
acquit_noment = data[data["CT_ACTION_TAKEN"].isin(["A", "D", "E", "T", "Y"])].reset_index(drop = True)

### filters for juvenile sanctions
juvie = data[data["CT_ACTION_TAKEN"] == "K"].reset_index(drop = True)

print("Total minors:", data["OFF_TRAN_NO"].nunique())
print("Minors acquitted (including for mental health reasons):" , acquit["OFF_TRAN_NO"].nunique())
print("Minors acquitted (excluding for mental health reasons):" , acquit_noment["OFF_TRAN_NO"].nunique())
print("Minors found delinquent or imposed juvenile sanctions):" , juvie["OFF_TRAN_NO"].nunique())
```
