Preparing dataset with basic information of each defendant.

### Importing libraries

```python
import pandas as pd
import numpy as np
```

### Importing dataset

```python
data_types = {
    "id": str,
    "County": str,
    "birth_date": str,
    "Init_Arrest_Dt": str,
    "Clerk_Dt": str,
    "CaseNumber": str,
    "Offense_Dt": str,
    "Capias_Dt": str,
    "Rearrest_Dt": str
}
```

```python
obts_basic = pd.read_csv("data/obts/obts_basic.csv", dtype = data_types)
obts_basic.columns = obts_basic.columns.str.upper()

obts_basic.info()
```

```python
obts_basic
```

### Formating dates

#### Controlling for null values

```python
obts_basic["BIRTH_DATE"] = obts_basic["BIRTH_DATE"].fillna("000000").replace("NA", "000000")

obts_basic["INIT_ARREST_DT"] = obts_basic["INIT_ARREST_DT"].fillna("00000000").replace("NA", "00000000")
obts_basic["CLERK_DT"] = obts_basic["CLERK_DT"].fillna("00000000").replace("NA", "00000000")
obts_basic["OFFENSE_DT"] = obts_basic["OFFENSE_DT"].fillna("00000000").replace("NA", "00000000")
obts_basic["CAPIAS_DT"] = obts_basic["CAPIAS_DT"].fillna("00000000").replace("NA", "00000000")
obts_basic["REARREST_DT"] = obts_basic["REARREST_DT"].fillna("00000000").replace("NA", "00000000")

obts_basic
```

#### Defining methods to convert to date-time format

```python
min_date = pd.to_datetime("NaT")  

def birth_convert(date_str):
    try:
        return pd.to_datetime(date_str, format = "%Y%m").strftime("%Y-%m")
    except Exception:
        return min_date
    
def date_convert(value):
    if value == "00000000":
        return min_date
    else:
        try:
            return pd.to_datetime(value, format = "%Y%m%d")
        except ValueError:
            return min_date
```

#### Converting to date-time format

```python
obts_basic["BIRTH_DATE"] = obts_basic["BIRTH_DATE"].apply(birth_convert)
print("Birth date formating complete.")
obts_basic["INIT_ARREST_DT"] = obts_basic["INIT_ARREST_DT"].apply(date_convert)
print("Initial arrest date formating complete.")
obts_basic["CLERK_DT"] = obts_basic["CLERK_DT"].apply(date_convert)
print("Clerk date formating complete.")
obts_basic["OFFENSE_DT"] = obts_basic["OFFENSE_DT"].apply(date_convert)
print("Offense date formating complete.")
obts_basic["CAPIAS_DT"] = obts_basic["CAPIAS_DT"].apply(date_convert)
print("Capias date formating complete.")
obts_basic["REARREST_DT"] = obts_basic["REARREST_DT"].apply(date_convert)
print("Rearrest date formating complete.")

obts_basic
```

### Exporting dataframe

```python
obts_basic.to_csv("data/obts/obts_basic_revised.csv", index = False)
```
