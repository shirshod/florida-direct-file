Merging basic, prosecutor, court and sentences databases for analysis

### Importing libraries

```python
import pandas as pd
```

### Importing datasets

```python
jr = pd.read_csv("data/obts/obts_basic_jr.csv")
sr = pd.read_csv("data/obts/obts_basic_sr.csv")

prosecutor = pd.read_csv("data/obts/obts_prosecutor.csv")
prosecutor.columns = prosecutor.columns.str.upper()

court = pd.read_csv("data/obts/obts_court.csv")
court.columns = court.columns.str.upper()

sentence_main = pd.read_csv("data/obts/sentence_main.csv")
sentence_other = pd.read_csv("data/obts/sentence_other.csv")
sentence_master = pd.read_csv("data/obts/sentence_master.csv")
```

### Merging for analysis on fixed duration sentences

#### Minors dataframe

```python
minors = pd.merge(jr, prosecutor, on = "ID", how = "left")
minors = pd.merge(minors, court, on = "ID", how = "left")
minors = pd.merge(minors, sentence_main, on = "ID", how = "left")

minors.info()
```

```python
### drops irrelevant rows
minors = minors.dropna(subset = ["ST_DAYS"]).reset_index(drop = True)
minors = minors[minors["COURT_DESIGNATOR"] != "J"].reset_index(drop = True)

minors
```

#### Adults dataframe

```python
adults = pd.merge(sr, prosecutor, on = "ID", how = "left")
adults = pd.merge(adults, court, on = "ID", how = "left")
adults = pd.merge(adults, sentence_main, on = "ID", how = "left")

adults.info()
```

```python
### drops irrelevant rows
adults = adults.dropna(subset = ["ST_DAYS"]).reset_index(drop = True)

adults
```

#### Exporting dataframes

```python
minors.to_csv("data/obts/minors_analysissentences.csv", index = False)
adults.to_csv("data/obts/adults_analysissentences.csv", index = False)
```

### Merging for analysis on life, death and juvenile sentences

#### Minors dataframe

```python
minors = pd.merge(jr, prosecutor, on = "ID", how = "left")
minors = pd.merge(minors, court, on = "ID", how = "left")
minors = pd.merge(minors, sentence_other, on = "ID", how = "left")

minors = minors.dropna(subset = ["ST_CONFINE"]).reset_index(drop = True)
minors = minors[minors["COURT_DESIGNATOR"] != "J"].reset_index(drop = True)

minors
```

#### Adults dataframe

```python
adults = pd.merge(sr, prosecutor, on = "ID", how = "left")
adults = pd.merge(adults, court, on = "ID", how = "left")
adults = pd.merge(adults, sentence_other, on = "ID", how = "left")

adults = adults.dropna(subset = ["ST_CONFINE"]).reset_index(drop = True)

adults
```

#### Exporting dataframes

```python
minors.to_csv("data/obts/minors_othersentences.csv", index = False)
adults.to_csv("data/obts/adults_othersentences.csv", index = False)
```

```python
minors = pd.merge(jr, prosecutor, on = "ID", how = "left")
minors = pd.merge(minors, court, on = "ID", how = "left")
minors = pd.merge(minors, sentence_master, on = "ID", how = "left")

minors = minors[minors["COURT_DESIGNATOR"] != "J"].reset_index(drop = True)

minors
```
